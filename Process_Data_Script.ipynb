```python
import re
import csv

def process_project_data(filename):
    filepath = "../" + filename
    print("In here")
    file1 = open(filename, 'r')
    Lines = file1.readlines()
    
    algorithm = "Q-Learning"
    alphaList = []
    gammaList = []
    energyList = []
    totalMigrationsList = []
    overUtilizedMigrationList = []
    pdmList = []
    slatahList = []
    slavList = []
    esvList = []
    runList = []

    alpha = 0.2
    gamma = 0.2
    energy = 0
    totalMigration = 0
    overMigrations = 0
    pdm = 0
    slatah = 0
    slav = 0
    esv = 0
    
    runCounter = 0
    # Strips the newline character
    for line in Lines:
        

        #IF THIS IS TRUE WE ARE IN THE FIRST PART OF THE NEW EXPERIMENT SO LETS CHANGE THE VALUES 
        if "Experiment name:" in line:
            runCounter += 1
            runList.append(runCounter)
            
            remainder = runCounter % 30
            
            alphaList.append(alpha)
            gammaList.append(gamma)

            if(remainder == 0):
                runCounter = 0
                
                print(alpha)
                print(gamma)
                if(gamma==1.0 and alpha==1.0):
                    continue
                elif(gamma < 1.0):
                    gamma = gamma + 0.2
                else:
                    gamma = 0.2 
                    alpha = alpha + 0.2 
                    


        elif "Energy consumption:" in line:
                    
            values = [float(s) for s in re.findall(r'-?\d+\.?\d*', line)]
            energy = values[0]
            energyList.append(energy)
            
        elif "Number of VM migrations:" in line:
        
            values = [float(s) for s in re.findall(r'-?\d+\.?\d*', line)]
            totalMigration = values[0]
            totalMigrationsList.append(totalMigration)
            
        elif "Number of VM migrations - Overutilized Hosts:" in line:
        
            values = [float(s) for s in re.findall(r'-?\d+\.?\d*', line)]
            overMigration = values[0]
            overUtilizedMigrationList.append(overMigration)

        elif "SLA perf degradation due to migration:" in line:
        
            values = [float(s) for s in re.findall(r'-?\d+\.?\d*', line)]
            pdm = values[0]/ 100
            pdmList.append(pdm)
            
        elif "SLA time per active host:" in line:
        
            values = [float(s) for s in re.findall(r'-?\d+\.?\d*', line)]
            slatah = values[0]/ 100
            slatahList.append(slatah)
            
        elif "SLAV:" in line:
        
            values = [float(s) for s in re.findall(r'-?\d+\.?\d*', line)]
            slav = values[0]
            slavList.append(slav)
            esv = energy * slav
            esvList.append(esv)

            
    
    #NOW LETS DO A CHECK TO MAKE SURE ALL THE LISTS ARE THE SAME SIZE 
    
    print(len(gammaList))
    print(len(alphaList))
    
    
    if(len(energyList) == len(totalMigrationsList) == len(overUtilizedMigrationList) == len(pdmList) == len(slatahList) == len(esvList) == len(alphaList) == len(gammaList)):
        print("They all match")
        
        #for i in range(len(energyList)):
            #print("%s %f %f %d %f %d %d %f %f %f" % (algorithm, alphaList[i], gammaList[i], runList[i],  energyList[i], totalMigrationsList[i], overUtilizedMigrationList[i], pdmList[i], slatahList[i], esvList[i]))
        
        with open('QLearning_Softmax.csv','w', newline='') as out:
            csv_out=csv.writer(out)
            csv_out.writerow(['Algorithm','Alpha','Gamma', 'Run', 'Energy Consumption', 'Total Migrations', 'Over-Utilized Migrations', 'PDM', 'SLATAH', 'SLAV', 'ESV'])
            
            for i in range(len(energyList)):
                data = [algorithm, alphaList[i], gammaList[i], runList[i],  energyList[i], totalMigrationsList[i], overUtilizedMigrationList[i], pdmList[i], slatahList[i], slavList[i], esvList[i]]
                csv_out.writerow(data)
```

```python
process_project_data("QLearning_Softmax_Results")
```
